```python
# make music data like mapping of type of sound to the sound (librosa loaded data)
musicData = [[]]
```

```python
#make data set
def addElement(path,_type="train"):
    global musicData
    y, sr = librosa.load(path)  # Load the audio file
    print(f"####Loaded {path} with sample rate {sr} and type {_type}")
    musicData.append([_type,y])
```

```python
def load_data_set():
    print("Loading music dataset...")
    global musicData
    for file in glob.glob("./Dataset/*/*"):
        _type = file.split("/")[-2]  # Get the type from the path
        addElement(file, _type)
    musicData = musicData[1:]  # Remove the first empty element    
    return musicData    


load_data_set()
```

```python


print("Music Data Loaded")
print(f"Total files in dataset: {len(musicData)}")
print(musicData)  
# play the first audio file in the dataset
label, y = musicData[0]
sr = 22050  # default librosa sample rate unless you changed it
ipd.display(ipd.Audio(y, rate=sr))

```
